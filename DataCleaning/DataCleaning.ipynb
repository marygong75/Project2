```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
```

```python
csv_file = "../DataCleanUp/Resources/Reveal_EEO1_for_2016.csv"
Gender_Job_df = pd.read_csv(csv_file)
Gender_Job_df.head()
```

```python
#Gender_Job_df[~Gender_Job_df.job_category.str.contains("Executives")]

# Get names of indexes for which column job_category = Totals 
indexNames = Gender_Job_df[Gender_Job_df['job_category'] == 'Totals'].index
# Delete these row indexes from dataFrame
Gender_Job_df.drop(indexNames , inplace=True)

# do the same for job_category = Previous_totals
indexNames = Gender_Job_df[Gender_Job_df['job_category'] == 'Previous_totals'].index
Gender_Job_df.drop(indexNames , inplace=True)

# Drop 'race' column & 'year' column
Gender_Job_df.drop(Gender_Job_df.columns[[2]], axis=1, inplace=True)
Gender_Job_df.drop(Gender_Job_df.columns[[1]], axis=1, inplace=True)




Gender_Job_df
```

```python
# Drop rows with 'NaN'
Gender_Job_df = Gender_Job_df.dropna()

Gender_Job_df
```
